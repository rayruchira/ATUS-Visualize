1. Top and Bottom 10:
    1. Avg: Time (non-zero and all), Happiness, Meaningfulness
    2. Variance: Time (non-zero and all), Happiness, Meaningfulness
2. Sideways Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
activity_time = pd.read_csv('activity_time_2021_.csv')
wb = pd.read_csv('wb_2021_.csv')

# Merge the datasets on the Activity Code/TRCODE
merged_data = pd.merge(activity_time, wb, left_on='Activity Code', right_on='TRCODE')

# Define a function to get top 5 and bottom 5
def get_top_bottom_5(data, column):
    sorted_data = data.sort_values(by=column, ascending=False)
    sorted_data = sorted_data.dropna(subset=[column])
    # sorted_data = sorted_data[sorted_data[column] != 0]

    top_5 = sorted_data.head(10)
    bottom_5 = sorted_data.tail(10)
    return top_5, bottom_5

# Define the columns to plot
columns_time = {
    'Average Time Spent': 'Average Time Spent',
    'Mean_NZ': 'Mean_NZ',
    'Variance': 'Variance',
    'Variance Non-Zero': 'Variance_NZ'
}
columns_wb = {
    'Happiness': 'Mean_WUHAPPY',
    'Meaningfulness': 'Mean_WUMEANING',
    'Happiness Variance': 'var_WUHAPPY',
    'Meaningfulness Variance': 'var_WUMEANING'
}
# Plot each metric
for title, column in columns_time.items():
    top_5, bottom_5 = get_top_bottom_5(activity_time, column)
    
    fig, axs = plt.subplots(1, 2, figsize=(15, 10))
    
    # Top 5 plot
    axs[0].bar(top_5['Activity'], top_5[column], color='b' if 'Variance' not in title else 'r')
    axs[0].set_title(f'Top 10 {title}')
    axs[0].set_xticklabels(top_5['Activity'], rotation=90)
    
    # Bottom 5 plot
    axs[1].bar(bottom_5['Activity'], bottom_5[column], color='b' if 'Variance' not in title else 'r')
    axs[1].set_title(f'Bottom 10 {title}')
    axs[1].set_xticklabels(bottom_5['Activity'], rotation=90)
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

# Plot each metric
for title, column in columns_wb.items():
    top_5, bottom_5 = get_top_bottom_5(wb, column)
    
    fig, axs = plt.subplots(1, 2, figsize=(15, 10))
    
    # Top 5 plot
    axs[0].bar(top_5['Activity'], top_5[column], color='b' if 'Variance' not in title else 'r')
    axs[0].set_title(f'Top 10 {title}')
    axs[0].set_xticklabels(top_5['Activity'], rotation=90)
    
    # Bottom 5 plot
    axs[1].bar(bottom_5['Activity'], bottom_5[column], color='b' if 'Variance' not in title else 'r')
    axs[1].set_title(f'Bottom 10 {title}')
    axs[1].set_xticklabels(bottom_5['Activity'], rotation=90)
    
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd


# Function to add rank columns
def add_rank_columns(df):
    for col in df.columns:
        rank_col_name = f'{col}_rank'  # New column name for rank
        df[rank_col_name] = df[col].rank(method='min', ascending=False)

# Add rank columns for activity_time
add_rank_columns(activity_time)

# Add rank columns for wb
add_rank_columns(wb)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

mpl.rcParams['font.family'] = 'Times New Roman'

# Assuming df_sorted is your dataframe containing the 'Activity' and 'Average Time Spent' columns
# and mean_scores_df containing the 'Activity' and 'Mean Score' columns.
df_sorted = activity_time
mean_scores_df = wb

# Filter out unwanted categories
df_sorted = df_sorted[df_sorted['Activity'] != 'work, main job']
mean_scores_df = mean_scores_df[mean_scores_df['Activity'] != 'work, main job']

# Sort the dataframes
df_sorted = df_sorted.sort_values(by='Average Time Spent_rank', ascending=False)
mean_scores_df = mean_scores_df.sort_values(by='Mean_WUHAPPY_rank', ascending=False)

# Merge the dataframes based on Activity and ATUS Supercategory
merged_df = pd.merge(mean_scores_df, df_sorted, left_on='Activity', right_on='Activity', how='inner')

common_activities = merged_df['Activity'].unique()

with open("/Users/ruchiraray/Documents/UT Austin/Soc-HHR/BEHAVIOR1kData/survey_analysis/test_ATUS", "r") as fp:  
    unique_categories = json.load(fp)

# Convert the list of unique categories to lowercase and strip trailing spaces
common_activities = [cat.lower().strip() for cat in unique_categories]

df_sorted["Activity"] = df_sorted["Activity"].str.lower().str.strip()
df_sorted = df_sorted[df_sorted['Activity'].isin(common_activities)]

mean_scores_df["Activity"] = mean_scores_df["Activity"].str.lower().str.strip()
mean_scores_df = mean_scores_df[mean_scores_df['Activity'].isin(common_activities)]

# Create a gradient color for each bar based on the sorted order
color_gradient_time = [plt.cm.Reds(i / len(df_sorted['Activity'])) for i in range(len(df_sorted['Activity']))]
color_gradient_score = [plt.cm.Reds(i / len(mean_scores_df['Activity'])) for i in range(len(mean_scores_df['Activity']))]

# Start the plot
fig, ax = plt.subplots(figsize=(24, 18))

# Plot the 'Average Time Spent' bars with a label for the legend
bars_time = ax.barh(df_sorted['Activity'], -df_sorted['Average Time Spent_rank'], color=color_gradient_time, edgecolor='black', label='Average Time Spent_rank')

# Overlay the 'Mean Score' bars with a label for the legend
bars_score = ax.barh(mean_scores_df['Activity'], mean_scores_df['Mean_WUHAPPY_rank'], color=color_gradient_score, edgecolor='black', label='Happy')

# Set the labels, titles, and ticks
ax.set_xlabel('Time / Happy', fontsize=28)
ax.set_yticklabels(df_sorted['Activity'], rotation=0, fontsize=24)
# ax.set_xlim(-10, 40)
# ax.set_xticks(range(-10, 41, 5))
ax.tick_params(axis='both', labelsize=20)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adjust layout and add legends
fig.tight_layout()
ax.legend(loc='upper right', fontsize=20)

# Save the plot to PDF
# plt.savefig("comparison_plot.pdf")

# Show the plot
plt.show()
```
